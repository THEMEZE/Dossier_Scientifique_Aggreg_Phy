```python
import numpy as np
import matplotlib.pyplot as plt

# Exemple de fonction n_tilde(x), tu peux changer
def n_tilde(x):
    return np.sin(x)  # exemple : gaussienne

# Domaine d'intégration
x = np.linspace(-5, 5, 1000)

def delta_rel(xs, x, n_tilde):
    n_x = n_tilde(x)
    n_sym = n_tilde(2*xs - x)
    
    num = np.trapz(((n_x - n_sym)/2)**2, x)
    den = np.trapz((n_x + n_sym)**2, x)
    den = np.trapz((n_x)**2, x)
    #den = 1
    
    return num / den

# Exemple d'utilisation
xs_values = np.linspace(-20, 20, 1000)
delta_values = [delta_rel(xs, x, n_tilde) for xs in xs_values]

# Visualisation
plt.plot(xs_values, delta_values)
plt.xlabel("$x_s$")
plt.ylabel("$\\delta_{rel}(x_s)$")
plt.show()
```
