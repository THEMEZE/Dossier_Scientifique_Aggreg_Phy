```python
!pip install pandas

import matplotlib.pyplot as plt
import pandas as pd

# Activer LaTeX
plt.rcParams["text.usetex"] = True
plt.rcParams["font.size"] = 16

# Lecture du CSV
df0 = pd.read_csv("./data/hydro10.csv", sep=";")

print(df0.head())

# Convertir les virgules en points
df0["x"] = df0["x"].str.replace(",", ".").astype(float)
df0["y"] = df0["y"].str.replace(",", ".").astype(float)

# ⚠️ Redéfinir x et y après conversion !
x0 = df0["x"]
y0 = df0["y"]

df1 = pd.read_csv("./data/hydro14.csv", sep=";")

print(df1.head())

# Convertir les virgules en points
df1["x"] = df1["x"].str.replace(",", ".").astype(float)
df1["y"] = df1["y"].str.replace(",", ".").astype(float)

# ⚠️ Redéfinir x et y après conversion !
x1 = df1["x"]
y1 = df1["y"]

df2 = pd.read_csv("./data/hydro18.csv", sep=";")

print(df1.head())

# Convertir les virgules en points
df2["x"] = df2["x"].str.replace(",", ".").astype(float)
df2["y"] = df2["y"].str.replace(",", ".").astype(float)

# ⚠️ Redéfinir x et y après conversion !
x2 = df2["x"]
y2 = df2["y"]



# Trace
plt.figure(figsize=(10,6))
# Exemple de couleurs agréables
colors = ["#1f77b4", "#ff7f0e", "#2ca02c"]  # bleu, orange, vert
colors = ["#636EFA", "#EF553B", "#00CC96"]  # bleu, rouge orangé, vert vif
colors = ["#FF6F61", "#6B5B95", "#88B04B"]  # corail, violet, vert pomme
colors = [ "#7F004F", "#FF5B69", "#57659E",]


#colors = ["#1A1A1D", "#C3073F", "#6F2232"]  # gris foncé, rouge vif, bordeaux



plt.plot(x0, y0, linewidth=4, color=colors[0], alpha=0.8, label=r"$\tau = 10 \, ms$")
plt.plot(x1, y1, linewidth=4, color=colors[1], alpha=0.8, label=r"$\tau = 14 \, ms$")
plt.plot(x2, y2, linewidth=4, color=colors[2], alpha=0.8, label=r"$\tau = 18 \, ms$")

plt.xlabel(r"$x/t ~ (\mu m /ms ) $", fontsize=30)
plt.ylabel(r"$n\;(\mu m ^{-1})$", fontsize=30)
#plt.title(r"$\tau = 17\,\mathrm{ms},\ \ell = 43\,\mu\mathrm{m}$", fontsize=22)
# Légende centrée en bas du graphe
plt.legend(
    fontsize=30,
    loc="upper left",
    #bbox_to_anchor=(0, -0.05),   # décale sous le graphique
    #frameon=False,        # pas de fond ni bordure
    borderpad=0.2,          # padding interne nul
    borderaxespad=1,      # marge avec les axes = 0
    #handlelength=0,       # longueur du trait de légende
    #handletextpad=0       # espace entre trait et texte
)

# Ticks plus gros
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)

#plt.ylim(-5.2, 22)

plt.grid(True)
plt.tight_layout()
plt.savefig("hydro_0.pdf", dpi=300, bbox_inches="tight")  # PDF pour article
plt.show()
```
