```julia
## fonctions de base

#packages

using Pkg
#Pkg.add("LinearAlgebra")
#Pkg.add("LinearSolve")
#Pkg.add("Plots")
#Pkg.add("Printf")
#Pkg.add("LaTeXStrings")
#Pkg.add("Interpolations")
#Pkg.add("NPZ")


using LinearAlgebra
using LinearSolve
using Plots
using LaTeXStrings
using Interpolations
using NPZ
using Printf


#-------------------------------------------------------------------------------------
#	dressing of function f (given as array f_discr, occupation ratio given as n_discr)
#-------------------------------------------------------------------------------------


function f_varphimat(lam_discr,gb)
    L = size(lam_discr)[1]
    varphi(lam) = 2*gb/( gb^2 + lam^2 )
    varphimat = zeros(L,L)
    

    for i = 1:L
        for j = 1:i
            varphimat[i,j] = varphi( lam_discr[i]-lam_discr[j] )
            varphimat[j,i] = varphi( lam_discr[i]-lam_discr[j] )
        end
    end
    varphimat = Symmetric(varphimat)
    return varphimat
end 

function f_dlam(lam_discr)
    L = size(lam_discr)[1]
    dlam = zeros(L)
    for i = 2:L-1
        dlam[i] = 0.5*(lam_discr[i+1]-lam_discr[i-1])
    end
    dlam[1] = 0.5*(lam_discr[2]-lam_discr[1])
    dlam[L] = 0.5*(lam_discr[L]-lam_discr[L-1])
    return dlam
end
    
function dress(gb, lam_discr, n_discr, f_discr)
    
    
    varphimat = f_varphimat(lam_discr,gb)
    dlam = f_dlam(lam_discr)
    
    A = I - 0.5/pi * varphimat*Diagonal(n_discr)*Diagonal(dlam)
    A\f_discr
end

#-------------------------------------------------------------------------------------
#	to evaluate charge density associated with function f
#-------------------------------------------------------------------------------------

function charge_density(gb, lam_discr, n_discr, f_discr)
    L = size(lam_discr)[1]
    varphi(lam) = 2*gb/( gb^2 + lam^2 )
    varphimat = zeros(L,L)
    dlam = zeros(L)

    for i = 1:L
        for j = 1:i
            varphimat[i,j] = varphi( lam_discr[i]-lam_discr[j] )
            varphimat[j,i] = varphi( lam_discr[i]-lam_discr[j] )
        end
    end
    varphimat = Symmetric(varphimat)

    for i = 2:L-1
        dlam[i] = 0.5*(lam_discr[i+1]-lam_discr[i-1])
    end
    dlam[1] = 0.5*(lam_discr[2]-lam_discr[1])
    dlam[L] = 0.5*(lam_discr[L]-lam_discr[L-1])
    A = I - 0.5/pi * varphimat*Diagonal(n_discr)*Diagonal(dlam)
    0.5/pi * dot(dlam , Diagonal(n_discr) * (A \ f_discr) )
end

#-------------------------------------------------------------------------
#	solve Yang-Yang equation
#-------------------------------------------------------------------------

function fun1(z)
    if z>0
        return log(1. +exp(-z))
    else
        return log(1. +exp(z)) - z
    end
end

function fun2(z)
    if z<0
        return 1. /(1. +exp(z))
    else
        return exp(-z)/(1. +exp(-z))
    end
end

function yangyang(gb, beta, lam_discr)
    L = size(lam_discr)[1]
    varphi(lam) = 2*gb/( gb^2 + lam^2 )
    varphimat = zeros(L,L)
    dlam = zeros(L)

    for i = 1:L
        for j = 1:i
            varphimat[i,j] = varphi( lam_discr[i]-lam_discr[j] )
            varphimat[j,i] = varphi( lam_discr[i]-lam_discr[j] )
        end
    end
    varphimat = Symmetric(varphimat)

    for i = 2:L-1
        dlam[i] = 0.5*(lam_discr[i+1]-lam_discr[i-1])
    end
    dlam[1] = 0.5*(lam_discr[2]-lam_discr[1])
    dlam[L] = 0.5*(lam_discr[L]-lam_discr[L-1])

    bare_E = 0.5 * beta[3] * lam_discr.^2 + beta[2] * lam_discr + beta[1] * ones(L)
    eps = copy(bare_E)
    n = fun2.(eps)

    #iterative solution
    diff = 1.
    while diff>1e-12
        old_n = copy(n)
        eps = bare_E - 0.5/pi * varphimat*Diagonal( fun1.(eps) )*dlam
        n = fun2.(eps)
        diff = norm(n-old_n)
    end
    eps = bare_E - 0.5/pi * varphimat*Diagonal( fun1.(eps) )*dlam
    n = fun2.(eps)
    return n
end

#-------------------------------------------------------------------------------------
#	compute effective velocity
#-------------------------------------------------------------------------------------

function veff(gb, lam_discr, n_discr)
    L = size(lam_discr)[1]
    varphi(lam) = 2*gb/( gb^2 + lam^2 )
    varphimat = zeros(L,L)
    dlam = zeros(L)

    for i = 1:L
        for j = 1:i
            varphimat[i,j] = varphi( lam_discr[i]-lam_discr[j] )
            varphimat[j,i] = varphi( lam_discr[i]-lam_discr[j] )
        end
    end
    varphimat = Symmetric(varphimat)

    for i = 2:L-1
        dlam[i] = 0.5*(lam_discr[i+1]-lam_discr[i-1])
    end
    dlam[1] = 0.5*(lam_discr[2]-lam_discr[1])
    dlam[L] = 0.5*(lam_discr[L]-lam_discr[L-1])
    dresser = I - 0.5/pi * varphimat*Diagonal(n_discr)*Diagonal(dlam)
    onedr = dresser\ones(L)
    iddr = dresser\copy(lam_discr)
    iddr ./ onedr
end

#-------------------------------------------------------------------------------------
#	time evolution for box expansion
#-------------------------------------------------------------------------------------

function evol_box_expansion(dt, x1, x2, gb, theta_tab, n_fun)
    npts_integral = 600
    L = size(x1)[1]
    
    veff1 = zeros(L)
    veff2 = zeros(L)
    #println(@sprintf "size x2 = %.3f , size theta_tab  = %.3f " size(x1)[1] size(theta_tab)[1])
    #display(x2)
    #display(theta_tab)
    f2 = linear_interpolation(x2, theta_tab)
    f1 = linear_interpolation(x1, theta_tab)
    
    f2 = interpolate((x2,), theta_tab, Gridded(Linear()))
    f2 = extrapolate(f2, Linear()) 
    
    f1 = interpolate((x1,), theta_tab, Gridded(Linear()))
    f1 = extrapolate(f1, Linear())
    #velocity of first and last points is the one of free particles
    veff1[1] = theta_tab[1]
    veff2[L] = theta_tab[L]

    for j in 2:L
        thet1 = theta_tab[j]
        if x1[j]<x2[1]
            thet2 = theta_tab[1]
        else
            thet2 = f2(x1[j])
        end
        lam_discr = LinRange(thet2,thet1,npts_integral)
        n_discr = n_fun.(lam_discr)
        veff1[j] = last(veff(gb, lam_discr, n_discr))
        #velocities of right contour (using parity symmetry)
        #veff2[L+1-j] = -veff1[j]
    end
    # sans symetrie
    for j in 1:L-1
        thet1 = theta_tab[j]
        if x2[j] > x1[L]
            thet2 = theta_tab[L]
        else 
            thet2 = f1(x2[j])
        end 
        lam_discr = LinRange(thet1, thet2, 600)
        n_discr = n_fun.(lam_discr)
        veff2[j] = veff(gb, lam_discr, n_discr)[1]
    end
    
    return x1 + dt*veff1, x2 + dt*veff2
end

#-------------------------------------------------------------------------------------
#	box expansion: calculate density (careful: return density in units of rapidities
#    for density in mu^-1 multiply by mass/hbar)
#-------------------------------------------------------------------------------------

function eval_density(x1, x2, gb, theta_tab, n_fun, ratio_m_hbar)
    L = size(x1)[1]
    
    dens = zeros(2,2*L)
    f2 = linear_interpolation(x2, theta_tab)
    f2 = interpolate((x2,), theta_tab , Gridded(Linear()))
    f2 = extrapolate(f2, Linear())
    f1 = linear_interpolation(x1, theta_tab)
    f1 = interpolate((x1,), theta_tab , Gridded(Linear()))
    f1 = extrapolate(f1, Linear())
    #density of first and last points is zero
    dens[1,1] = x1[1]
    dens[2,1] = 0
    dens[1,L+1] = x2[L]
    dens[2,L+1] = 0
    #println( @sprintf "n_p = %.3f" ratio_m_hbar * charge_density(gb, theta_tab, n_fun.(theta_tab), ones(size(theta_tab)[1]) ))
    for j in 2:L
        thet1 = theta_tab[j]
        if x1[j]<x2[1]
            thet2 = theta_tab[1]
        else
            #println("nnn")
            thet2 = f2(x1[j])
        end
        lam_discr = LinRange(thet2,thet1,600)
        n_discr = n_fun.(lam_discr)
        dens[1,j] = x1[j]
        dens[2,j] = ratio_m_hbar * charge_density(gb, lam_discr, n_discr, ones(size(lam_discr)[1]) )
        #if j == L 
        #    println( @sprintf " j = %.3f , theta 1 = %.3f , theta 2 = %.3f , n_p1 = %.3f" j thet2 thet1 dens[2,j])
        #end 
        #velocities of right contour (using parity symmetry)
        #dens[1,L+j] = x2[L+1-j]      
        #dens[2,L+j] = dens[2,j]
    end
    #without symmetry 
    for j in 1:L-1
        thet1 = theta_tab[j]
        if x2[j]>x1[L]
            thet2 = theta_tab[L]
        else 
            #println("nnn")
            thet2 = f1(x2[j])
        end 
        lam_discr = LinRange(thet1,thet2,600)
        n_discr = n_fun.(lam_discr)
        dens[1,2*L+1-j] = x2[j] 
        dens[2,2*L+1-j] = ratio_m_hbar * charge_density(gb, lam_discr, n_discr, ones(size(lam_discr)[1]) ) 
        #if j == 1 
        #    println( @sprintf " j = %.3f , theta 1 = %.3f , theta 2 = %.3f , n_p1 = %.3f" j thet1 thet2 dens[2,2*L+1-j])
        #end 
    end
    #sort densities according to positions, and return
    dens[:,sortperm(dens[1, :])]
end

#------------------------------------------------------------------------------------------------
#	function for simulating box expansion. Writes results in file 'data/density_expansion_t.npd'
#     where 't' is the time (takes a list of times to be saved as argument)
#------------------------------------------------------------------------------------------------

function box_expansion(n_fun, dt, list_t_save, theta_max, npts, gb, ratio_m_hbar)
    tmax = maximum(list_t_save)
    theta_tab = LinRange(-theta_max, theta_max, npts)
    x1 = -0.5 * ones(npts) + 0.0000001*theta_tab
    x2 = 0.5 * ones(npts) + 0.0000001*theta_tab
        
    t = 0
    while t<=tmax
        x1, x2 = evol_box_expansion(dt, x1, x2, gb, theta_tab, n_fun)
        t += dt
        
        for ts in list_t_save
            if t-dt<ts && ts<=t
                mydensity = eval_density(x1, x2, gb, theta_tab, n_fun, ratio_m_hbar)
                println(@sprintf "t = %.3f saved" t)
                #filename = @sprintf "%s/density_expansion_n_theta_etoile_conv_py_%.3f_%.3f_%.3f_%.3f.npz" new_date Float64(mu) Float64(T) Float64(Taille) Float64(t)
                #filename = @sprintf "%s/density_expansion_n_py_%.3f_%.3f_%.3f_%.3f.npz" new_date Float64(mu) Float64(T) Float64(Taille) Float64(t)
                filename = @sprintf "%s/density_expansion_n_theta_etoile_py_%.3f_%.3f_%.3f_%.3f.npz" new_date Float64(mup/kB) Float64(T) Float64(Taille) Float64(t)
                #filename = @sprintf "%s/density_expansion_n_tout_py_%.3f_%.3f_%.3f_%.3f.npz" new_date Float64(mup/kB) Float64(T) Float64(Taille) Float64(t)
                npzwrite(filename, mydensity)
            end
        end
    end
end

```

```julia
function DeltaMat(theta_discr , gbar )
    L = size(theta_discr)[1]
    varphi(theta) = 2*gbar/( gbar^2 + theta^2 )
    varphimat = zeros(L,L)
    

    for i = 1:L
        for j = 1:i
            varphimat[i,j] = varphi( theta_discr[i]-theta_discr[j] )
            varphimat[j,i] = varphi( theta_discr[i]-theta_discr[j] )
        end
    end
    varphimat = Symmetric(varphimat)
    return varphimat
end 

function Delta(theta_discr)
    L = size(theta_discr)[1]
    dtheta = zeros(L)
        
    for i = 2:L-1
        dtheta[i] = 0.5*(theta_discr[i+1]-theta_discr[i-1])
    end
    dtheta[1] = 0.5*(theta_discr[2]-theta_discr[1])
    dtheta[L] = 0.5*(theta_discr[L]-theta_discr[L-1])
    
    return dtheta
end 

function montrer_les_rho(theta_discr, nu_discr)
    L = length(theta_discr)
    
    varphimat = DeltaMat(theta_discr , gbar)
    dtheta = Delta(theta_discr)
    
    # Plot nu_tilde
    p = plot(xlabel="\$\\tilde{\\theta}\$", ylabel="f", legend=:topright)
    plot!(p , theta_discr , nu_discr, label="\$\\tilde{\\nu}\$")
    # Méthode 0
    rho_s_0 = 0.5 / pi * dress(gbar , theta_discr, nu_discr, ones(L))
    rho_0 = nu_discr .* rho_s_0
    
    # Méthode 1
    Mat_F_1 = I - (1 / (2 * pi)) * Diagonal(dtheta) * Diagonal(nu_discr) * varphimat
    B = (1 / (2 * pi)) * nu_discr
    rho_1 = Mat_F_1 \ B
    rho_s_1 = rho_1 ./ nu_discr

    # Méthode 2
    Mat_F_2 = I - 0.5 / pi * varphimat * Diagonal(nu_discr) * Diagonal(dtheta)
    B = (1 / (2 * pi)) * ones(L)
    rho_s_2 = Mat_F_2 \ B
    rho_2 = nu_discr .* rho_s_2
    
    # Plot résultats de la méthode 0
    plot!(p, theta_discr, rho_s_0, label="Meth 0 : \$\\tilde{\\rho_s}= \\frac{1}{2\\pi}\\\$")
    plot!(p, theta_discr, rho_0, label="Meth 0 : \$\\tilde{\\rho} = \\tilde{\\nu} * \\tilde{\\rho_s}\$")
    
    # Plot résultats de la méthode 1
    plot!(p, theta_discr, rho_1, label="Meth 1 : \$\\tilde{\\rho}= \\frac{\\tilde{\\nu}}{2\\pi} ( 1 + \\tilde{\\Delta} * \\tilde{\\rho}) \$")
    plot!(p, theta_discr, rho_s_1, label="Meth 1 : \$\\tilde{\\rho}_s  = \\frac{\\tilde{\\rho}}{\\tilde{\\nu}}  \$")
    
    # Plot résultats de la méthode 2
    plot!(p, theta_discr, rho_s_2, label="Meth 2 : \$\\tilde{\\rho_s}= \\frac{1}{2\\pi}\$")
    plot!(p, theta_discr, rho_2, label="Meth 2 : \$\\tilde{\\rho} = \\tilde{\\nu} * \\tilde{\\rho_s}\$")

    display(p)
    
    #p1 = plot(xlabel="\$\\tilde{\\theta}\$", ylabel="\$\\Delta\$", legend=:topright)

    # Plot des différences entre les résultats des deux méthodes
    #plot(p1,theta_discr, rho_s_1 - rho_s_2, label="\$ \\Delta \\tilde{\\rho_s}\$")
    #plot(p1,theta_discr, rho_1 - rho_2, label="\$\\Delta \\tilde{\\rho}\$")

    # Légendes et labels
    #display(p1)
end

#-------------------------------------------------------------------------------------
#	theta start
#-------------------------------------------------------------------------------------
#Pkg.add("Roots")
using Roots

function find_nu_theta_star_discr( theta_discr , theta_star  , nu_discr )
    return ifelse.(theta_discr .> theta_star, 0, nu_discr)
end 

function find_theta_star(theta_discr, nu_discr ,  gbar , v)
    
    # Définition de la fonction f_veff_star_theta_star
    f_veff_star_theta_star = theta_star -> linear_interpolation(theta_discr, veff( gbar, theta_discr, find_nu_theta_star_discr(theta_discr, theta_star, nu_discr)))(theta_star)-v
    
    return fzero(f_veff_star_theta_star, theta_discr[1] , last(theta_discr))
end
```

```julia
using Plots
using Printf
using Interpolations

# Définir les constantes
hbar = 1.05457182e-25   # um^2.kg/ms
mass =  1.44e-25        # kg (mass of Rubidium 87)
kB = 1.380649e-26       # um^2.ms^-2.kg.nK^{-1}
a3D = 5.3e-3            # um
om_perp = 2 * pi * 2.56  # kHz (transverse frequency)

# Calculer les constantes dérivées
g = 2 * hbar * a3D * om_perp        # effective 1d repulsion strength
c = mass / hbar^2 * g              # um^{-1}
gbar = g / hbar                   # um.ms^{-1}
```

```julia
using Printf
using FileIO  # Pour la fonction npzwrite (Assurez-vous d'avoir installé cette bibliothèque)

# Fonction qui charge les paramètres et effectue les vérifications de la "boule"
function check_file_and_parameters(filename_save::String, param_cible::Array{Float64}, delta::Float64, norme::Int64)
    # Vérification si le fichier existe
    if isfile(filename_save)
        # Ouvre le fichier et charge les paramètres
        try
            param_list = npzread(filename_save)  # Charge le contenu du fichier .npz
        catch e
            println("Erreur lors du chargement du fichier : $e")
            return false
        end
        
        L = size(param_cible)[1]
        # Vérification si l'un des paramètres dans `param_list` correspond à une "boule" selon la norme donnée
        for param_set in param_list
            # Calcul de la norme entre les paramètres d'entrée et les paramètres du fichier
            if size(param_set[1]) == L 

                if est_un_tableau_non_vide_non_nule(param_cible) == false 
                    return false
                end 
                # On ajoute 0.000001 à chaque élément de param_cible qui est nul
                param_cible_m = param_cible .+ (param_cible .== 0) * 0.000001
                distance = calculate_distance((param_cible_m .- param_set)./param_cible_m norme)

                # On vérifie si la distance est inférieure à un seuil donné (par exemple delta * 100)
                if distance <= delta 
                    println("Boule trouvée avec les paramètres $param_set")
                    return param_set
                end
            end
        end
        return false
    else
        println("Le fichier n'existe pas.")
        return false
    end
end

function calculate_distance(x::Array{Float64}, norme::Int64)
    # Calcul de la distance selon la norme donnée
    
    if isempty(x)
        return 0.0  # Si le vecteur est vide, on retourne une distance nulle
    end

    # Norme 0 : Comptage des éléments non nuls
    #if norme == 0
    #   return sum(x .!= 0)  # Norme 0 : nombre de valeurs non nulles dans x

    # Norme 0 : 0 si x est nul, inf sinon
    if norme == 0
        return 0.0 if all(x .== 0) else Inf  # Si tous les éléments de x sont nuls, retourne 0, sinon Inf
    
    
    # Norme 1 : Somme des valeurs absolues des éléments
    elseif norme == 1
        return sum(abs.(x))  # Norme 1 : somme des valeurs absolues
    
    # Norme 2 : Racine carrée de la somme des carrés des éléments
    elseif norme == 2
        return sqrt(sum(x .^ 2))  # Norme 2 : racine de la somme des carrés
    
    # Norme infinie : Maximum des valeurs absolues des éléments
    elseif norme == Inf
        return maximum(abs.(x))  # Norme infinie : maximum des différences

    # Norme p (p positif) : Somme des valeurs élevées à la puissance p, puis racine p-ème
    elseif norme > 0 && isnumeric(norme)
        return (sum(abs.(x) .^ norme)) ^ (1 / norme)  # Norme p : somme des différences à la puissance p, puis racine p-ème

    else
        println("Norme inconnue ou invalide.")
        return Inf  # Retourne une valeur infinie si la norme est inconnue ou invalide
    end
end

# Exemple de fonction pour générer un nom de fichier basé sur les paramètres
function generate_filename(filename0::String, param_set::Array{Float64})
    # Initialisation du nom de fichier avec le nom de base
    filename = filename0
    
    # Itération sur les éléments de param_set et ajout à filename
    for i in 1:size(param_set, 1)  # On itère sur chaque ligne de param_set (s'il s'agit d'une matrice)
        # Ajout de chaque paramètre au nom du fichier en formatant à 3 chiffres après la virgule
        filename *= @sprintf("_%.3f", param_set[i])
    end
    
    # Ajout du suffixe ".npz" à la fin du nom de fichier
    filename *= ".npz"
    
    return filename
end

function est_un_tableau_non_vide_non_nule(x::::Array{Float64})
    if isa(x, Array) && !isempty(x) && all(x .!= 0)
        println("x est un tableau non vide avec tous les éléments non nuls.")
        return true 
    elseif x == false
    println("x est égal à false.")
    else
        println("x n'est ni un tableau valide, ni false.")
    end
end

function simul_dans_boule_existe(filename_save::String, filename0::String, param_cible::Array{Float64}, delta::Float64, norme::Int64)

    # forma 
    forma = ["n_p" "mu" "T" "t_deform" "x0" "t_exp" "Nat" "Taille" ]
    if isfile(filename_save)
        param_set = check_file_and_parameters(filename_save, param_cible, delta, norme) 
        
        if param_set != false 
            filename = generate_filename(filename0, param_set) 
            try
                out_list = npzread(filename)  # Charge le contenu du fichier .npz
                return out_list
            catch e
                println("Erreur lors du chargement du fichier : $e")
            return false
        end 
    else
        println("Le fichier n'existe pas.")
        npzwrite(filename_save, [] )
        println("saved : ", filename_save)
        return false
    end
end

function list_exist(filename::String)

    try
        out_list = npzread(filename_save)  # Charge le contenu du fichier .npz
        return out_list
    catch e
        println("Erreur lors du chargement du fichier : $e")
        return false
    end

end

function param_list_exist(filename::String)
    param_list = list_exist(filename)
    if param_list == false 
        return []
    end
    return param_list
    
end
```

## evolution d'un bord 

```julia
#-------------------------------------------------------------------------------------
#	time evolution for box expansion
#-------------------------------------------------------------------------------------

function evol_edge_expansion(dt, x, gb, theta_tab, n_fun)
    npts_integral = 600
    L = size(x)[1]
    
    veff_tab = zeros(L)

    f = linear_interpolation(x, theta_tab)

    #velocity of first and last points is the one of free particles
    veff_tab[1] = theta_tab[1]

    for j in 2:L
        thet1 = theta_tab[j]
        thet2 = theta_tab[1]
        lam_discr = LinRange(thet2,thet1,npts_integral)
        n_discr = n_fun.(lam_discr)
        #println("vell")
        #@time veff_tab[j] = last(veff(gb, lam_discr, n_discr))
        veff_tab[j] = last(veff(gb, lam_discr, n_discr))
        #velocities of right contour (using parity symmetry)
    end
    
    return x + dt*veff_tab
end

function evol_edge_expansion_balistic(dt, x, gb, theta_tab, n_fun)
    npts_integral = 600
    L = size(x)[1]
    
    veff_tab = zeros(L)

    f = linear_interpolation(x, theta_tab)

    #velocity of first and last points is the one of free particles
    veff_tab[1] = theta_tab[1]

    for j in 2:L
        thet1 = theta_tab[j]
        thet2 = theta_tab[1]
        lam_discr = LinRange(thet2,thet1,npts_integral)
        n_discr = n_fun.(lam_discr)
        veff_tab[j] = last(veff(gb, lam_discr, n_discr))
        #velocities of right contour (using parity symmetry)
    end
    
    return x + dt*veff_tab
end

#-------------------------------------------------------------------------------------
#	box expansion: calculate density (careful: return density in units of rapidities
#    for density in mu^-1 multiply by mass/hbar)
#-------------------------------------------------------------------------------------

function eval_edge_density(x, gb, theta_tab, nu_fun, ratio_m_hbar)
    npts_integral = 600
    
    L = size(x)[1]
    dens = zeros(2,L)
    dens[1,1] = x[1]
    dens[2,1] = 0
    thet1 = theta_tab[1]


    for j in 2:L
        thet1 = theta_tab[j]
        thet2 = theta_tab[1]
        lam_discr = LinRange(thet2,thet1,npts_integral)
        nu_discr = nu_fun.(lam_discr)
        dens[1,j] = x[j]      
        dens[2,j] = ratio_m_hbar * charge_density(gb, lam_discr, nu_discr, ones(size(lam_discr)[1]) )
    end
    
    println( "la densité haute est : " , dens[2,L])

    #sort densities according to positions, and return
    return dens[:,sortperm(dens[1, :])] #, nu
end

```

```julia

function model_prel_mu(T,mu)
    #println(@sprintf "T = %.3f ,mu = %.3f " T mu)
    
    theta_discr_integrer = LinRange(-theta_max, theta_max, 600)
    mup = mu * kB
    beta = [-mup / (kB * T) 0 mass /(kB * T)]
    nu_discr = yangyang(gbar, beta, theta_discr)
    nu_fun = LinearInterpolation(theta_discr, nu_discr)
    thet2 , thet1 , npts_integral = theta_discr_integrer[1], theta_discr_integrer[600] , 600
    lam_discr = LinRange(thet2,thet1,npts_integral)
    nu_discr = nu_fun.(lam_discr)    
    densite = mass/hbar* charge_density(gbar, lam_discr, nu_discr, ones(size(lam_discr)[1]) )
    #println(densite)
    return densite
end

function model_mu(T,mu)
    
    OutDensite = 0*ones(size(T)[1])
    for i in 1:size(T)[1]
        #println(@sprintf "i =, T = %.3f ,mu = %.3f " i T[i] mu[i])
        OutDensite[i] = model_prel_mu(T[i],mu[i])
        #println("OutDensite = " ,OutDensite )
    end
    
    if 1 == 0                                                                    
        # Création du graphique
        p = plot(xlabel="mu", ylabel="densite", legend=:topright)
        # Ajout des points avec des marqueurs personnalisés
        plot!(p, mu, OutDensite, seriestype=:scatter, markersize=12, markerstrokewidth=2, markerstrokecolor=:DarkSlateGrey)
        # Affichage du graphique
        display(p)
    end
    
    #println(OutDensite)
    return OutDensite
    
end

if 1 == 0 
    T = LinRange(10, 600 , 3)
    Densite = 56.46504685*ones(size(T)[1])
    p0 = 50*ones(size(T)[1])
    fit = curve_fit(model_mu, T , Densite , p0 )
    
    mu = coef(fit)
    # Création du graphique
    p = plot(xlabel="mu", ylabel="densite", legend=:topright)
    # Ajout des points avec des marqueurs personnalisés
    plot!(p, mu, model_mu(T,mu), seriestype=:scatter, markersize=12, markerstrokewidth=2, markerstrokecolor=:DarkSlateGrey)
    # Affichage du graphique
    display(p)
    
    # Création du graphique
    p = plot(xlabel="T", ylabel="mu", legend=:topright)
    # Ajout des points avec des marqueurs personnalisés
    plot!(p, T, mu, seriestype=:scatter, markersize=12, markerstrokewidth=2, markerstrokecolor=:DarkSlateGrey)
    # Affichage du graphique
    display(p)
end 

function trier_x_et_y(x::Vector{T}, y::Vector{T}) where T
    # Obtenir les indices qui trient x en ordre croissant
    indices_tries = sortperm(x)
    
    # Trier x en ordre croissant
    x_tries = x[indices_tries]
    
    # Réorganiser y en conséquence
    y_tries = y[indices_tries]
    
    return x_tries, y_tries
end
```

```julia
function charge_density(gb, lam_discr, n_discr, f_discr)
    L = size(lam_discr)[1]
    varphi(lam) = 2*gb/( gb^2 + lam^2 )
    varphimat = zeros(L,L)
    dlam = zeros(L)

    for i = 1:L
        for j = 1:i
            varphimat[i,j] = varphi( lam_discr[i]-lam_discr[j] )
            varphimat[j,i] = varphi( lam_discr[i]-lam_discr[j] )
        end
    end
    varphimat = Symmetric(varphimat)

    for i = 2:L-1
        dlam[i] = 0.5*(lam_discr[i+1]-lam_discr[i-1])
    end
    dlam[1] = 0.5*(lam_discr[2]-lam_discr[1])
    dlam[L] = 0.5*(lam_discr[L]-lam_discr[L-1])
    A = I - 0.5/pi * varphimat*Diagonal(n_discr)*Diagonal(dlam)
    0.5/pi * dot(dlam , Diagonal(n_discr) * (A \ f_discr) )
end

#-------------------------------------------------------------------------
#	solve Yang-Yang equation
#-------------------------------------------------------------------------

function fun1(z)
    if z>0
        return log(1. +exp(-z))
    else
        return log(1. +exp(z)) - z
    end
end

function fun2(z)
    if z<0
        return 1. /(1. +exp(z))
    else
        return exp(-z)/(1. +exp(-z))
    end
end

function yangyang(gb, beta, lam_discr)
    L = size(lam_discr)[1]
    varphi(lam) = 2*gb/( gb^2 + lam^2 )
    varphimat = zeros(L,L)
    dlam = zeros(L)

    for i = 1:L
        for j = 1:i
            varphimat[i,j] = varphi( lam_discr[i]-lam_discr[j] )
            varphimat[j,i] = varphi( lam_discr[i]-lam_discr[j] )
        end
    end
    varphimat = Symmetric(varphimat)

    for i = 2:L-1
        dlam[i] = 0.5*(lam_discr[i+1]-lam_discr[i-1])
    end
    dlam[1] = 0.5*(lam_discr[2]-lam_discr[1])
    dlam[L] = 0.5*(lam_discr[L]-lam_discr[L-1])

    bare_E = 0.5 * beta[3] * lam_discr.^2 + beta[2] * lam_discr + beta[1] * ones(L)
    eps = copy(bare_E)
    n = fun2.(eps)

    #iterative solution
    diff = 1.
    while diff>1e-12
        old_n = copy(n)
        eps = bare_E - 0.5/pi * varphimat*Diagonal( fun1.(eps) )*dlam
        n = fun2.(eps)
        diff = norm(n-old_n)
    end
    eps = bare_E - 0.5/pi * varphimat*Diagonal( fun1.(eps) )*dlam
    n = fun2.(eps)
    return n
end

# Définir les constantes
hbar = 1.05457182e-25   # um^2.kg/ms
mass =  1.44e-25        # kg (mass of Rubidium 87)
kB = 1.380649e-26       # um^2.ms^-2.kg.nK^{-1}
a3D = 5.3e-3            # um
om_perp = 2 * pi * 2.56  # kHz (transverse frequency)

# Calculer les constantes dérivées
g = 2 * hbar * a3D * om_perp        # effective 1d repulsion strength
c = mass / hbar^2 * g              # um^{-1}
gbar = g / hbar      


function model_prel_mu(T,mu)
    #println(@sprintf "T = %.3f ,mu = %.3f " T mu)
    
    theta_discr_integrer = LinRange(-theta_max, theta_max, 600)
    mup = mu * kB
    beta = [-mup / (kB * T) 0 mass /(kB * T)]
    nu_discr = yangyang(gbar, beta, theta_discr)
    nu_fun = LinearInterpolation(theta_discr, nu_discr)
    thet2 , thet1 , npts_integral = theta_discr_integrer[1], theta_discr_integrer[600] , 600
    lam_discr = LinRange(thet2,thet1,npts_integral)
    nu_discr = nu_fun.(lam_discr)    
    densite = mass/hbar* charge_density(gbar, lam_discr, nu_discr, ones(size(lam_discr)[1]) )
    #println(densite)
    return densite
end

function model_mu(T,mu)
    
    OutDensite = 0*ones(size(T)[1])
    for i in 1:size(T)[1]
        #println(@sprintf "i =, T = %.3f ,mu = %.3f " i T[i] mu[i])
        OutDensite[i] = model_prel_mu(T[i],mu[i])
        #println("OutDensite = " ,OutDensite )
    end
    
    if 1 == 0                                                                    
        # Création du graphique
        p = plot(xlabel="mu", ylabel="densite", legend=:topright)
        # Ajout des points avec des marqueurs personnalisés
        plot!(p, mu, OutDensite, seriestype=:scatter, markersize=12, markerstrokewidth=2, markerstrokecolor=:DarkSlateGrey)
        # Affichage du graphique
        display(p)
    end
    
    #println(OutDensite)
    return OutDensite
    
end

mu = coef(curve_fit(model_mu, T*ones(1) , n_p*ones(1) , mu*kB/hbar*ones(1) ))[1]  
```

```julia
using Interpolations
using Plots

function calculate_max_epsilon_with_plot(observed_x, observed_y, simulated_x, simulated_y)
    """
    Calcule la plus grande valeur de epsilon qui permet à simulated_y de rester dans le tube défini par observed_y ± epsilon,
    et trace le graphique avec les courbes et le remplissage de l'écart.

    Args:
        observed_x (array-like): Les abscisses des données observées.
        observed_y (array-like): Les ordonnées des données observées.
        simulated_x (array-like): Les abscisses des valeurs simulées.
        simulated_y (array-like): Les ordonnées des valeurs simulées.

    Returns:
        float: La plus petite valeur de epsilon qui satisfait la condition.
    """

    # Vérification si observed_x est croissant
    if !issorted(observed_x)
        # Si ce n'est pas le cas, inverser les valeurs de observed_x et observed_y
        observed_x = reverse(observed_x)
        observed_y = reverse(observed_y)
        simulated_x = reverse(simulated_x)
        simulated_y = reverse(simulated_y)
    end
    
    # Convertir les entrées en tableaux Julia (Vector)
    observed_x = observed_x
    observed_y = observed_y
    simulated_x = simulated_x
    simulated_y = simulated_y
    
    # Vérifier que les données ne sont pas vides
    if length(observed_x) == 0
        return nothing
    end
    
    # Créer une interpolation linéaire des valeurs simulées sur les abscisses des données observées
    interpolated_simulated_y = interpolate((simulated_x,), simulated_y, Gridded(Linear()))
    
    # Interpoler les données simulées aux abscisses observées
    interpolated_simulated_y_at_observed_x = [interpolated_simulated_y(x) for x in observed_x]
    
    # Calculer la plus grande valeur de epsilon nécessaire pour que simulated_y soit dans le tube
    epsilon = maximum(abs.(interpolated_simulated_y_at_observed_x .- observed_y))
    
    plot_f = false  # Contrôle de l'affichage du graphique
    if plot_f
        # Tracer les données observées, simulées et le tube
        p = plot(observed_x, observed_y, label="Observed Data", color=:blue, linewidth=2)
        
        # Tracer la courbe des données simulées
        plot!(simulated_x, simulated_y, label="Simulated Data", color=:green, linewidth=2)
        
        # Tracer la courbe des données simulées interpolées
        plot!(observed_x, interpolated_simulated_y_at_observed_x, label="Interpolated Simulated Data", color=:orange, linewidth=2)
        
        # Remplir le tube entre observed_y - epsilon et observed_y + epsilon
        plot!(observed_x, observed_y .- epsilon, observed_x, observed_y .+ epsilon, fillrange=0, fillalpha=0.3, label="Tube (±ε)", color=:gray)
        
        # Ajouter des titres et des labels
        title!("Comparison of Observed and Simulated Data with ε Tube")
        xlabel!("X")
        ylabel!("Y")
        grid!(:on)
    end

    # Retourner la valeur de epsilon
    return epsilon
end
```

```julia
function sort_and_unique(x, y)
    # Trier x et obtenir les indices triés
    sorted_indices = sortperm(x)

    # Trier x en utilisant ces indices
    sorted_x = x[sorted_indices]
    sorted_y = y[sorted_indices]

    # Créer une liste pour les valeurs uniques et les indices associés
    unique_x = []
    unique_y = []

    # Remplir les listes uniques en gardant la première occurrence
    for i in 1:length(sorted_x)
        if isempty(unique_x) || sorted_x[i] != unique_x[end]
            push!(unique_x, sorted_x[i])
            push!(unique_y, sorted_y[i])
        end
    end

    # Retourner les résultats uniques
    return unique_x, unique_y
end
```

```julia
using Interpolations
using Plots

function calculate_chi_2_max(observed_x, observed_y, simulated_x, simulated_y; p=2, plot_flag=false)
    """
    Calcule la plus grande valeur de epsilon qui permet à simulated_y de rester dans le tube défini par observed_y ± epsilon,
    et trace le graphique avec les courbes et le remplissage de l'écart.

    Args:
        observed_x (Array): Les abscisses des données observées.
        observed_y (Array): Les ordonnées des données observées.
        simulated_x (Array): Les abscisses des valeurs simulées.
        simulated_y (Array): Les ordonnées des valeurs simulées.
        p (Number): Paramètre de la norme (par défaut 2).
        plot_flag (Bool): Si vrai, trace les graphiques.

    Returns:
        Array: La plus petite valeur de epsilon qui satisfait la condition.
    """
    # Vérifier que les données ne sont pas vides et que les tailles des vecteurs sont compatibles
    if isempty(observed_x) || isempty(observed_y) || isempty(simulated_x) || isempty(simulated_y)
        return nothing
    end
    
    if length(observed_x) != length(observed_y)
        println("Erreur : les dimensions de observed_x et observed_y doivent être égales.")
        return nothing
    end

    if length(simulated_x) != length(simulated_y)
        println("Erreur : les dimensions de simulated_x et simulated_y doivent être égales.")
        return nothing
    end

    # Trier les données observées
    observed_x, observed_y = sort_and_unique(observed_x, observed_y)

    simulated_x, simulated_y = sort_and_unique(simulated_x, simulated_y)
    
    
    
    # Interpoler les valeurs simulées sur les abscisses des données observées
    interp = LinearInterpolation(simulated_x, simulated_y)
    interpolated_simulated_y = [interp(x) for x in observed_x]

    # Définir la norme
    norme = p == 0 ? (x -> Inf) :
            p == Inf ? (x -> maximum(abs.(x))) :
            (x -> sum(abs.(x) .^ p)^(1/p))

    # Calculer epsilon
    epsilon = norme((interpolated_simulated_y .- observed_y) ./ interpolated_simulated_y)
    epsilon0 = norme(interpolated_simulated_y .- observed_y)

    # Tracer si plot_flag est activé
    if plot_flag
        # Créer une nouvelle figure avec une taille personnalisée
        p1plot = plot(observed_x, observed_y, label="Observed Data", color=:blue, linewidth=2, figsize=(10, 6))
        plot!(p1plot, simulated_x, simulated_y, label="Simulated Data", color=:green, linewidth=2)
        plot!(p1plot, observed_x, interpolated_simulated_y, label="Interpolated Simulated Data", color=:orange, linewidth=2)

        # Remplir le tube entre observed_y - epsilon0 et observed_y + epsilon0
        plot!(p1plot, observed_x, observed_y .- epsilon0, fillrange=observed_y .+ epsilon0, fillalpha=0.3, color=:gray, label="Tube (±ε) ε₀=$(round(epsilon0, digits=3))")
        title!("Comparison of Observed and Simulated Data with ε Tube")
        xlabel!("X")
        ylabel!("Y")
        #legend(:topright)  # Affichage de la légende à droite en haut
        #grid!(:true)
        display(p1plot)

        # Tracer le second graphique pour l'écart normalisé
        p2plot = plot(observed_x, (interpolated_simulated_y .- observed_y) ./ interpolated_simulated_y,
             label="Pour norme", color=:orange, linewidth=2, figsize=(10, 6))
        plot!(p2plot, observed_x, -epsilon .* ones(length(observed_x)), fillrange=epsilon .* ones(length(observed_x)),
              fillalpha=0.3, color=:gray, label="Tube (±ε) ε=$(round(epsilon, digits=3))")
        title!("Comparison of Observed and Simulated Data with ε Tube")
        xlabel!("X")
        ylabel!("Y")
        #legend(:topright)  # Affichage de la légende à droite en haut
        #grid!(:true)
        display(p2plot)
    end

    println( "epsilon0^2 " , epsilon0^2 ) 

    # Retourner la valeur calculée d'epsilon0^2
    return epsilon0^2
end
```

```julia
using Optim
using Plots
#import Pkg; Pkg.add("CSV")
using CSV
#import Pkg; Pkg.add("DataFrames")
using DataFrames
using Dates
using Printf
using LineSearches  # Importer LineSearches pour accéder à FixedStepSize

# Définir le chemin et nom du fichier dynamiquement
modes = "T-x0"
filename = @sprintf "%s/Optimisation_%s_%s.csv" new_date modes Dates.format(Dates.now(), "yyyy-mm-dd-HH-MM-SS")
println("filename : " , filename)

# Initialisation d'un DataFrame vide pour le fichier CSV
df = DataFrame(T=Float64[], x0=Float64[], Sum=Float64[])
CSV.write(filename, df)  # Crée un fichier vide avec les colonnes

# Initialisation des tableaux pour accumuler les résultats dans la mémoire
XT = []
Yx0 = []
Sum = Float64[]

# Facteurs d'échelle
scale_T = 0.2  # Échelle pour T
scale_x0 = 2.0 # Échelle pour x0 (car 1 tolérance pour T = 0.5 tolérance pour x0)

# Créer un graphique vide avec des labels
plotly()  # Permet un graphique interactif
popty = plot3d(xlabel="T", ylabel="x0", zlabel="norme2^2", linewidth=2, color=:blue, legend=:topright)

# Fonction pour sauvegarder les résultats dans le fichier CSV
function append_results_to_csv(filename, T, x0, summ)
    # Charger le DataFrame existant
    df = CSV.read(filename, DataFrame)
    
    # Ajouter une nouvelle ligne avec les résultats
    new_row = DataFrame(T=[T], x0=[x0], Sum=[summ])
    df = vcat(df, new_row)
    
    # Sauvegarder le DataFrame mis à jour
    CSV.write(filename, df)
end

# Fonction objective
# Fonction objective
function objective_function(p_scaled , n_p, Nat, temp_deform, Temps_exp , mode , filename)
    # Décodage des variables mises à l'échelle
    println("p_scaled, n_p, Nat, temp_deform, Temps_exp, mode, filename: ", p_scaled, ", ", n_p, ", ", Nat, ", ", temp_deform, ", ", Temps_exp, ", ", mode, ", ", filename)

    T = p_scaled[1] * scale_T
    x0 =  p_scaled[2] * scale_x0
    println("Début de l'optimisation avec les paramètres : ", T ," , " ,  x0 )
    

    # Calcul de la fonction modèle
    y_model = model_fit_inhomo_deux(x_data_m, [T, x0] , n_p, Nat, temp_deform, Temps_exp , mode )
    if any(isnan, y_model)
        return Inf
    end

    # Calcul de la somme des erreurs quadratiques
    #summ = sum((y_data_m .- y_model).^2)
    #summ = sum((y_smoothed .- y_model).^2)#/sum((y_smoothed).^2)
    #summ = calculate_max_epsilon_with_plot(x_data_m, y_smoothed, x_data_m, y_model)
    summ = calculate_chi_2_max(reverse(x_data_m), reverse(y_smoothed), x_data_m, y_model; p=Inf, plot_flag=true)
     
    println("Somme des erreurs quadratiques : ", summ)
    
    # Ajouter la somme à la liste Sum
    push!(XT ,T)
    push!(Yx0, x0)
    push!(Sum, summ)
    
    # Mise à jour du graphique à chaque itération
    plot!(popty, XT, Yx0, Sum, linewidth=2, color=:blue)  # Met à jour le graphique avec les nouvelles données
    display(popty)

    # Enregistrer dans le fichier CSV
    append_results_to_csv(filename, T, x0, summ)
    
    return summ
end


# Initialisation des paramètres mis à l'échelle
#p0, lower_bounds , upper_bounds = [  325.088  / scale_T , 3 / scale_x0 ] , [  0.1 * p0[1] , -1 ] , [  1.5 * p0[1] , 7 ]  # Mise à l'échelle initiale
p0, lower_bounds , upper_bounds = [ 1800/ scale_T , 20/ scale_x0 ] , [  200/scale_T , -20/ scale_x0  ] , [  3000/scale_T , 20/ scale_x0  ]  # Mise à l'échelle initiale

#p0, lower_bounds , upper_bounds = [981.810 / scale_T] , [100/scale_T] , [2000/scale_T]  # Mise à l'échelle initiale
#p0  ,lower_bounds , upper_bounds =  [18.339 / scale_x0] , [-20 ] , [20]     # Mise à l'échelle initiale
#lower_bounds = [  0.1 * p0[1] , -1 ]
#upper_bounds = [  1.5 * p0[1] , 7 ]
#lower_bounds = 0.1 .* p0.*ones(size(p0)[1])
#upper_bounds = 3 .* p0.*ones(size(p0)[1])

# Options d'optimisation
options = Optim.Options(
    x_tol=1.0,       # Tolérance commune pour les variables mises à l'échelle
    f_tol=1e-8,      # Tolérance pour la fonction objective
    g_tol=1e-8,      # Tolérance pour le gradient
    show_trace=true  # Afficher les informations de progression
    #iterations=1000,    # Nombre d'itérations maximal
)

# Optimisation


f = x -> objective_function(x, n_p, Nat, temp_deform, Temps_exp, mode , filename)

run = true 
if run 
    result = optimize(
        f,  # Fonction lambda
        lower_bounds, 
        upper_bounds, 
        p0, 
        Fminbox(LBFGS(linesearch=LineSearches.HagerZhang())),
        #BFGS(), 
        options
    )
end 



# Décodage des résultats
p_optimized_scaled = Optim.minimizer(result)
T_optimized = p_optimized_scaled[1] * scale_T
x0_optimized = p_optimized_scaled[2] * scale_x0

println("Résultat de l'optimisation : T = $T_optimized, x0 = $x0_optimized")
```

```julia
# Définir le chemin predefinie et nom du fichier dynamiquement 
using Optim
using Plots
#import Pkg; Pkg.add("CSV")
using CSV
#import Pkg; Pkg.add("DataFrames")
using DataFrames
using Dates
using Printf
using LineSearches  # Importer LineSearches pour accéder à FixedStepSize
modes = "T-x0-liste"
filename = @sprintf "%s/Optimisation_%s_%s.csv" new_date modes Dates.format(Dates.now(), "yyyy-mm-dd-HH-MM-SS")
println("filename : " , filename)

# Initialisation d'un DataFrame vide pour le fichier CSV
df = DataFrame(T=Float64[], x0=Float64[], Sum=Float64[])
CSV.write(filename, df)  # Crée un fichier vide avec les colonnes

# Initialisation des tableaux pour accumuler les résultats dans la mémoire
XT = []
Yx0 = []
Sum = Float64[]

# Facteurs d'échelle
scale_T = 0.2  # Échelle pour T
scale_x0 = 2.0 # Échelle pour x0 (car 1 tolérance pour T = 0.5 tolérance pour x0)

# Créer un graphique vide avec des labels
plotly()  # Permet un graphique interactif
popty = plot3d(xlabel="T", ylabel="x0", zlabel="norme2^2", linewidth=2, color=:blue, legend=:topright)

f = x -> objective_function(x, n_p, Nat, temp_deform, Temps_exp, mode , filename)

# Création des listes avec les intervalles spécifiés
liste_T = range(550, stop=1800, length=5) |> collect  # 5 éléments de 100 à 2000 inclus
liste_x0 = range(10, stop=20, length=5) |> collect     # 4 éléments de 5 à 20 inclus

# Division élément par élément pour obtenir les listes "scaled"
liste_T_scaled = liste_T ./ scale_T
liste_x0_scaled = liste_x0 ./ scale_x0

# Affichage des résultats
println("liste_T_scaled: ", liste_T_scaled)
println("liste_x0_scaled: ", liste_x0_scaled)

# Boucle sur les éléments de liste_T_scaled et liste_x0_scaled
for T_scaled in liste_T_scaled
    for x0_scaled in liste_x0_scaled
        p0_scaled = [T_scaled, x0_scaled]
        f(p0_scaled)  # Appel de la fonction f avec p0_scaled
    end
end
        
```
