```julia
using Statistics, Random

Random.seed!(1234)

# Paramètres
N = 1_000_000           # taille d'une run
nblocks = 100       # nombre de blocs
mu = 1.0           # moyenne réelle
sigma = 2.0        # écart-type réel

# Génère 10 blocs indépendants de taille N
data = [randn(N) * sigma .+ mu for i in 1:nblocks]

# -- Méthode 1 : SE_N direct (sur une seule run de taille N)
sample1 = data[1]
SE_N_direct = std(sample1) / sqrt(N)

# -- Méthode 2 : SE_10N (tout concaténé)
all_data = reduce(vcat, data)
SE_10N = std(all_data) / sqrt(length(all_data))

# -- Méthode 3 : SE sur 10 blocs
means_blocks = [mean(block) for block in data]
sb = std(means_blocks)                # écart-type des moyennes de bloc
SE_10blocks = sb / sqrt(nblocks)

println("Comparaison des estimateurs :")
println("SE_N direct    ≈ ", SE_N_direct)
println("SE_10N         ≈ ", SE_10N, "   (attendu ~ SE_N/√10)")
println("SE_10 blocs    ≈ ", SE_10blocks, "   (attendu ~ SE_10N)")

println(((SE_N_direct/SE_10N)^2-nblocks)/nblocks , " " ,nblocks )
```
